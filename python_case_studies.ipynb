##Case Study 1 - Bank Account Manager

```python
class BankAccount:

  def __init__(self, owner, account_no, balance, deactivated = False):

    self.owner = owner
    self.account_no = account_no
    self.balance = balance
    self.deactivated = deactivated

  def check_balance(self):

    if self.deactivated:
      print("Account is deactivated")
      return
    else:
      print(f"Your current Balance is: {self.balance}")

  def deposit(self, amount):
    if self.deactivated:
      print("Account is deactivated!")
      return

    if amount <= 0:
      print("Invalid Amount")
      return

    else:
      self.balance += amount
      print(f"Current Balance: {self.balance}")

  def withdraw(self, amount):

    if self.deactivated:
      print("Account is deactivated!")
      return

    if amount <= 0:
      print("Invalid Amount")
      return

    elif amount <= self.balance:
      self.balance -= amount
      print(f"Current Balance: {self.balance}")

    else:
      print("Insufficient Balance")

  def deactivate(self):

    if self.balance == 0:
      print(f" Your Account (account number - {self.account_no}) has been Deactivated")
      self.deactivated = True
    else:
      print("Withdraw all the money to Deactivate account")

# dictionary to store customers
bank_db = {}

# user validation
def get_int(prompt):

  while True:

    try:
      return int(input(prompt))

    except ValueError:
      print("Invalid input!")

def get_float(prompt):

  while True:

    try:
      return float(input(prompt))

    except ValueError:
      print("Invalid amount!")

def create_account():

  name = input("Enter your name: ")
  account_no = get_int("Enter your account number: ")

  if account_no in bank_db:
    print("Account already Exists")
    return

  balance = get_float("Enter initial deposit: ")
  account = BankAccount(name, account_no, balance)
  bank_db[account_no] = account

  print("Account created Successfully")

def access_account():

  account_no = get_int("Enter your account number: ")

  if account_no not in bank_db:
    print("Account not available!")
    return

  account = bank_db[account_no]

  # operations
  while True:

    print()
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Deactivate")
    print("5. Exit")

    choice = int(input("Choose: "))

    if choice == 1:
      account.check_balance()

    elif choice == 2:
        amount = get_float("Enter amount: ")
        account.deposit(amount)

    elif choice == 3:
        amount = get_float("Enter amount: ")
        account.withdraw(amount)

    elif choice == 4:
        account.deactivate()

    elif choice == 5:
        break

    else:
        print("Invalid choice")

# main menu
while True:

  print("---- Welcome ----")
  print()
  print("1. Create Account")
  print("2. Access Account")
  print("3. Exit")
  print()

  choice = get_int(input("Choose: "))

  if choice == 1:
    create_account()

  elif choice == 2:
    access_account()

  elif choice == 3:
    print("Thank You")
    break

  else:
    print("Invalid Choice")


```

## Case Study 2 - Weather Data Analyzer

```python
# dictionary to store weather on multiple locations
weather_db = {}

# user validation
def get_int(prompt):

  while True:

    try:
      return int(input(prompt))

    except ValueError:
      print("Invalid input!")

def get_float(prompt):

  while True:

    try:
      return float(input(prompt))

    except ValueError:
      print("Invalid Input")

# weather classification
def weather_condition(temp, humidity, precipitation, wind):

  if precipitation > 0.5:
    return "Rainy"

  elif wind > 30:
    return "Windy"

  elif humidity > 80 and temp > 25:
    return "Humid"

  elif temp > 30 and humidity < 50:
    return "Sunny"

  elif humidity > 60 and precipitation == 0:
    return "Cloudy"

  else:
    return "Normal"

def add_weather():

  location = input("Enter Location: ").strip().lower()

  temperature = get_float("Enter temperature(Celsius): ")
  humidity = get_float("Enter humidity percentage: ")
  precipitation = get_float("Enter precipitation percentage: ")
  wind = get_float("Enter wind speed(km/h): ")

  # creates a empty list to store values for each locations
  if location not in weather_db:
      weather_db[location] = {
          "temperature": [],
          "humidity": [],
          "precipitation": [],
          "wind": []
      }

  weather_db[location]["temperature"].append(temperature)
  weather_db[location]["humidity"].append(humidity)
  weather_db[location]["precipitation"].append(precipitation)
  weather_db[location]["wind"].append(wind)

  print("Data added successfully!")
  print()

def analyze_weather():

  location = input("Enter Location: ").strip().lower()

  if location not in weather_db:
      print("No data found!")
      print()
      return

  data = weather_db[location]

  # taking last value in weather_db to get latest analysis
  temp = data["temperature"][-1]
  humidity = data["humidity"][-1]
  precipitation = data["precipitation"][-1]
  wind = data["wind"][-1]

  weather_now = weather_condition(temp, humidity, precipitation, wind)

  print("Today is a", weather_now, "day")
  print(f"Average temperature of {location} is", temp_avg(data["temperature"]))
  print(f"Average humidity of {location} is", humidity_avg(data["humidity"]))
  print()

def temp_avg(temperatures):

  if not temperatures:
    return 0

  else:
    average_temperature = sum(temperatures) / len(temperatures)
    return average_temperature

def humidity_avg(humiditys):

  if not humiditys:
    return 0

  else:
    average_humidity = sum(humiditys) / len(humiditys)
    return average_humidity

# Main Menu
while True:

    print("---- Weather System ----")
    print("1. Add Weather Data")
    print("2. Weather Analysis")
    print("3. Exit")
    print()

    choice = get_int("Choose: ")

    if choice == 1:
        add_weather()

    elif choice == 2:
        analyze_weather()

    elif choice == 3:
        print("Thank You")
        break

    else:
        print("Invalid choice")
```
